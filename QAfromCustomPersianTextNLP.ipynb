```python
import PyPDF2

file_path = "/content/drive/MyDrive/NLP/rules.pdf"
with open(file_path, "rb") as file:
    reader = PyPDF2.PdfReader(file)
    document_text = ""
    for page in reader.pages:
        document_text += page.extract_text()
```

```python
print(document_text)
```

```python
import torch
from transformers import T5ForConditionalGeneration, T5Tokenizer
import PyPDF2

# 1. مدل T5 و tokenizer را بارگیری کنید
model_name = "t5-small"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

# 2. آماده‌سازی ورودی برای مدل با اطلاعات زمینه‌ای بیشتر
question = "چه مقرراتی برای ماده ۱۶ قانون مدیریت خدمات کشوری وجود دارد؟"
contextual_info = "قانون مدیریت خدمات کشوری یکی از قوانین مهم در کشور است که شرایط و مقررات مختلفی را برای خدمات عمومی تعیین می‌کند."
input_prefix = f"پاسخ به سوال با اطلاعات زمینه‌ای: {contextual_info} متن: {question} "

# 3. تقسیم متن فایل به بخش‌های کوچکتر
max_length = 512 - len(tokenizer.encode(input_prefix))  # حداکثر طول مجاز
chunks = [document_text[i:i + max_length] for i in range(0, len(document_text), max_length)]

# 4. تنظیم پارامترهای تولید
temperature = 0.7  # کنترل خلاقیت مدل
top_k = 50         # محدود کردن مدل به 50 کلمه با احتمال بالاتر
top_p = 0.9        # محدود کردن انتخاب به 90% از مجموع احتمالات

# 5. تولید پاسخ برای هر بخش با فرمت ورودی بهبود یافته
responses = []
for chunk in chunks:
    input_text = input_prefix + chunk
    input_ids = tokenizer.encode(input_text, return_tensors="pt")

    # تولید پاسخ با استفاده از پارامترهای تنظیم شده
    with torch.no_grad():
        output_ids = model.generate(
            input_ids,
            max_length=200,
            temperature=temperature,
            top_k=top_k,
            top_p=top_p,
            do_sample=True  # فعال‌سازی نمونه‌گیری برای استفاده از top_k و top_p
        )

    # رمزگشایی پاسخ
    response = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    responses.append(response)

# 6. ترکیب پاسخ‌ها (یا مدیریت آن‌ها بر اساس نیاز)
final_response = " ".join(responses)
print(final_response)
```
